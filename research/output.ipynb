```python
import tkinter as tk
from sign_language_conversion import logger
from sign_language_conversion.pipeline.prediction import HandGestureRecognition

STAGE_NAME = "Prediction of Model"
try:
    logger.info("*******************")
    logger.info(f">>>>>> stage {STAGE_NAME} started <<<<<<")
    obj = HandGestureRecognition()
    obj.recognize_gesture()
    logger.info(f">>>>>> stage {STAGE_NAME} completed <<<<<<\n\nx==========x")
except Exception as e:
    logger.exception(e)
    raise e

class GestureInterface:
    
    def remove_repeated_letters():
    # Read the content of output.txt
        with open('output.txt', 'r') as file:
            content = file.read()

        # Function to remove repeated letters from a word
        def remove_repeats(word):
            new_word = []
            for char in word:
                if not new_word or char != new_word[-1]:
                    new_word.append(char)
            return ''.join(new_word)

        # Process each word in the content and remove extra spaces
        words = content.split()
        modified_words = [remove_repeats(word) for word in words]

        # Join the modified words back into a single string with a single space between each word
        modified_content = ' '.join(modified_words)

        # Write the modified content back to output.txt
        with open('output.txt', 'w') as file:
            file.write(modified_content)

    remove_repeated_letters()
    
    # def remove_repeated_letters():
    #     # Read the content of output.txt
    #     with open('output.txt', 'r') as file:
    #         content = file.read()

    #     # Function to remove repeated letters from a word
    #     def remove_repeats(word):
    #         new_word = []
    #         count = 1
    #         for i in range(1, len(word)):
    #             if word[i] == word[i - 1]:
    #                 count += 1
    #             else:
    #                 if count > 5:
    #                     new_word.append(word[i - 1] * count)
    #                 count = 1
    #         if count > 5:
    #             new_word.append(word[-1] * count)
    #         return ''.join(new_word)

    #     # Process each word in the content and remove extra spaces
    #     words = content.split()
    #     modified_words = [remove_repeats(word) for word in words]

    #     # Join the modified words back into a single string with a single space between each word
    #     modified_content = ' '.join(modified_words)

    #     # Write the modified content back to output.txt
    #     with open('output.txt', 'w') as file:
    #         file.write(modified_content)

    # # Call the function to process the file
    # remove_repeated_letters()



    def __init__(self, filename):
        self.root = tk.Tk()
        self.root.title("Hand Gesture Recognition")
        self.text_label = tk.Label(self.root, text="Recognized Gesture:", font=("Arial", 16))
        self.text_label.pack(pady=10)
        
        # Initial text for gesture_label
        self.gesture_label = tk.Label(self.root, text=self.fetch_letters_from_file(filename), font=("Arial", 20, "bold"))
        self.gesture_label.pack(pady=20)
        
        self.filename = filename
        self.update_gesture()

    def fetch_letters_from_file(self, file_path):
        try:
            # Open the file in read mode
            with open(file_path, 'r') as file:
                content = file.read()
            
            # Extract letters from the content
            letters = ''.join(filter(str.isalpha, content))
            
            return letters

        except FileNotFoundError:
            return "File not found."
        except Exception as e:
            return str(e)

    def preprocess_gesture(self, gesture):
        processed_gesture = ""
        prev_char = ""
        for char in gesture:
            if char != prev_char:
                processed_gesture += char
                prev_char = char
            elif char == ' ':
                processed_gesture += char
        return processed_gesture

    def update_gesture(self):
        try:
            with open(self.filename, 'r') as file:
                gesture = file.read()
                processed_gesture = self.preprocess_gesture(gesture)
            self.gesture_label.config(text=processed_gesture)
        except FileNotFoundError:
            pass  # Handle the case when the file is not found
        self.root.after(1000, self.update_gesture)  # Update gesture every second

    def run(self):
        self.root.mainloop()

if __name__ == "__main__":
    interface = GestureInterface("output.txt")
    interface.run()
```
